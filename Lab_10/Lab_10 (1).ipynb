```python
import os
import requests
```

```python
# -------------------------------
# 1Ô∏è‚É£ Accept filename and count character occurrences
# -------------------------------
def count_characters(filename):
    try:
        with open(filename, 'r', encoding='utf-8') as f:
            text = f.read()
        counts = {}
        for ch in text:
            counts[ch] = counts.get(ch, 0) + 1
        print("\nCharacter counts:")
        for ch, cnt in counts.items():
            print(f"'{ch}': {cnt}")
    except FileNotFoundError:
        print("File not found!")
```

```python

# -------------------------------
# 2Ô∏è‚É£ Count vowels and consonants
# -------------------------------
def count_vowels_consonants(filename):
    try:
        with open(filename, 'r', encoding='utf-8') as f:
            text = f.read().lower()
        vowels = "aeiou"
        v_count = sum(1 for ch in text if ch in vowels)
        c_count = sum(1 for ch in text if ch.isalpha() and ch not in vowels)
        print(f"\nVowels: {v_count}, Consonants: {c_count}")
    except FileNotFoundError:
        print("File not found!")
```

```python
# -------------------------------
# 3Ô∏è‚É£ Compute total size of all files in current directory
# -------------------------------
def total_file_size(path='.'):
    total_size = 0
    for filename in os.listdir(path):
        if os.path.isfile(os.path.join(path, filename)):
            total_size += os.path.getsize(os.path.join(path, filename))
    print(f"\nTotal size of all files in '{path}': {total_size} bytes")

```

```python
# -------------------------------
# 4Ô∏è‚É£ Generate quiz from question.txt and answer.txt
# -------------------------------
def generate_quiz(question_file, answer_file):
    try:
        with open(question_file, 'r', encoding='utf-8') as qf, \
             open(answer_file, 'r', encoding='utf-8') as af:
            questions = qf.readlines()
            answers = af.readlines()

        print("\nüìò Quiz Time! üìò\n")
        for i, q in enumerate(questions):
            print(f"Q{i+1}. {q.strip()}")
            user_ans = input("Your answer: ").strip()
            correct = answers[i].strip()
            if user_ans.lower() == correct.lower():
                print("‚úÖ Correct!\n")
            else:
                print(f"‚ùå Wrong! Correct answer: {correct}\n")
    except FileNotFoundError:
        print("One or both quiz files not found!")

```

```python
# -------------------------------
# 5Ô∏è‚É£ Fetch data from URL and write to a file
# -------------------------------
def fetch_and_save(url, output_file):
    try:
        response = requests.get(url)
        if response.status_code == 200:
            print("\nData fetched successfully!")
            print(response.text[:500])  # print first 500 chars
            with open(output_file, 'w', encoding='utf-8') as f:
                f.write(response.text)
            print(f"Data saved to '{output_file}'")
        else:
            print(f"Failed to fetch URL (status code: {response.status_code})")
    except Exception as e:
        print(f"Error fetching URL: {e}")
```

```python
# -------------------------------
# üöÄ MAIN EXECUTION (All Tasks)
# -------------------------------
def main():
    filename = "sample.txt"          # Change this to your text file name
    question_file = "question.txt"   # Quiz questions file
    answer_file = "answer.txt"       # Quiz answers file
    url = "https://example.com"      # URL to fetch
    output_file = "url_data.html"     # File to save fetched data

    count_characters(filename)
    count_vowels_consonants(filename)
    total_file_size()
    generate_quiz(question_file, answer_file)
    fetch_and_save(url, output_file)
    print("\n‚úÖ All tasks completed successfully!")


if __name__ == "__main__":
    main()
```
