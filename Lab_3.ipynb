```python
# WAP to claculate area of a triangle using Herons formula

a = int(input("Enter length of edge 1: "))
b = int(input("Enter length of edge 2: "))
c = int(input("Enter length of edge 3: "))

s = (a+b+c)/2
print("Area of triangle using Heron's formula : ", sqrt(s*(s-a)*(s-b)*(s-c)))
```

```python
# WAP a program to generate given bill, given its Quantity sold, price per item, Tax, Discount.

#   ************************BILL************************
#   Quantity                                        0000
#   Price Per Item                                  0000
#   Amount                                          0000
#   ----------------------------------------------------
#   Discount                                        0000
#   Discount Total                                  0000
#   Tax                                             0000
#   ----------------------------------------------------
#                                                   0000

Quantity = int(input("Enter Quantity: "))
PPI = float(input("Enter Price Per Item: "))
Discount = float(input("Enter Discount amount: "))
Tax = float(input("Enter Tax (%): "))


Amount = Quantity * PPI
DiscountTotal = Amount - Discount
TaxAmount = (DiscountTotal * Tax) / 100
FinalTotal = DiscountTotal + TaxAmount


print("\n  {:*^50}".format(" BILL "))
print("  {:<30}{:>20}".format("Quantity", f"{Quantity}"))
print("  {:<30}{:>20.2f}".format("Price Per Item", PPI))
print("  {:<30}{:>20.2f}".format("Amount", Amount))
print("  {:─<50}".format(""))
print("  {:<30}{:>20.2f}".format("Discount", Discount))
print("  {:<30}{:>20.2f}".format("Discount Total", DiscountTotal))
print("  {:<30}{:>19.2f}%".format("Tax", Tax))
print("  {:─<50}".format(""))
print("  {:>50.2f}".format(FinalTotal))
```

```python
#   WAP to calculate student's result based on two examination,
#   1 sport and 3 activities
#   conducted weightage of activities 30%
#   sports 20% and examination 50%
#   let MAX_MARKS = 100

#   WAP to calculate student's result based on two examination,
#   1 sport and 3 activities
#   conducted weightage of activities 30%
#   sports 20% and examination 50%
#   let MAX_MARKS = 100

print("\n  {:*^50}".format(" STUDENT RESULT CALCULATOR "))

m_sports = int(input("Enter sports marks (out of 100): "))
m_activities = int(input("Enter activities marks (out of 100): "))
m_exam = int(input("Enter examination marks (out of 100): "))

weighted_sports = m_sports * 0.20
weighted_activities = m_activities * 0.30
weighted_exam = m_exam * 0.50
total_weighted = weighted_sports + weighted_activities + weighted_exam

print("\n  {:<30}{:>20.2f}".format("Sports (20%)", weighted_sports))
print("  {:<30}{:>20.2f}".format("Activities (30%)", weighted_activities))
print("  {:<30}{:>20.2f}".format("Exams (50%)", weighted_exam))
print("  {:─<50}".format(""))
print("  {:<30}{:>20.2f}".format("Final Weighted Marks", total_weighted))
```

```python
#   WAP to convert degree Fahrenheit to Celsius

print("\n  {:*^50}".format(" FAHRENHEIT TO CELSIUS CONVERTER "))

f = float(input("Enter temperature in Fahrenheit: "))
c = (f - 32) * 5 / 9

print("  {:<30}{:>20.2f}°F".format("Fahrenheit", f))
print("  {:<30}{:>20.2f}°C".format("Celsius", c))
```

```python
#   WAP to calculate total amount of money in piggy bank,
#   given coins of Rs. 10, 5, 2, 1

print("\n  {:*^50}".format(" PIGGY BANK TOTAL CALCULATOR "))

coins_10 = int(input("Enter number of Rs.10 coins: "))
coins_5 = int(input("Enter number of Rs.5 coins: "))
coins_2 = int(input("Enter number of Rs.2 coins: "))
coins_1 = int(input("Enter number of Rs.1 coins: "))

total = (coins_10 * 10) + (coins_5 * 5) + (coins_2 * 2) + (coins_1 * 1)

print("\n  {:<30}{:>20}".format("Rs.10 Coins", coins_10))
print("  {:<30}{:>20}".format("Rs.5 Coins", coins_5))
print("  {:<30}{:>20}".format("Rs.2 Coins", coins_2))
print("  {:<30}{:>20}".format("Rs.1 Coins", coins_1))
print("  {:─<50}".format(""))
print("  {:<30}{:>20} ₹".format("Total Amount", total))
```
