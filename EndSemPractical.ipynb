### 1. Write a recursive Python function to compute the digital root of a number without using loops.

```python
def digital_root(n):
# Base Case:
# If the number has only one digit , that digit itself is the real root 
    if n<10:
        return n

# Recursive Case:
#1. n // 10 gives the number without the last digit 
#2. n%10 gives the last digit 
#3. digital_root(n // 10) recursively finds the sum of digits
#4. Add the last digit (n % 10) to that sum 
#5. Call digital_root again to ensure the result is a single digit 
    return digital_root(digital_root(n //10) + n % 10)


# Calling the function with input 22
print(digital_root(22))
```

### 2. Write a program to compute F(M,N), where F(M,N) can be recursively defined asF(M,N) =1 if m=0 or M&gt;N&gt;1 and F(M,N)=F(M-1,N)+F(M-1,N-1), otherwise

```python
def F(M,N):
    # Base case:
    # If M becomes 0
    # OR
    # If M > N and N > 1
    # then return 1 and stop further recursion 
    if M==0 or (M>N and N>1):
        return 1
    
    # Recursive case:
    # Call the function twice:
    # 1. Reduce M by 1, keep N same
    # 2. Reduce both M and N ny 1
    # Add the results of both recursive calls 
    return F(M-1,N) + F(M-1,N-1)


# Function call with M = 2 and N = 1
print(F(2,1))
```
