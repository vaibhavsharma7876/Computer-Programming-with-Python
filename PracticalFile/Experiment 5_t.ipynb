```python
# Printing number pattern using nested loops

for i in range(1, 6):
    for j in range(1, i + 1):
        print(j, end="")
    print()
```

```python
# Multiplication table of a number

n = int(input("Enter a number: "))

print("Multiplication Table of", n)
for i in range(1, 11):
    print(n, "x", i, "=", n * i)
```

```python
# Function using keyword arguments

def student_info(name, age, course):
    print("Student Information")
    print("Name:", name)
    print("Age:", age)
    print("Course:", course)

# Function call using keyword arguments in any order
student_info(course="Python Programming", name="Rahul", age=20)
```

```python
# Absolute value, square root and cube of a number

import math

num = float(input("Enter a number: "))

absolute_value = abs(num)
square_root = math.sqrt(abs(num))
cube = num ** 3

print("Absolute Value:", absolute_value)
print("Square Root:", square_root)
print("Cube:", cube)
```
