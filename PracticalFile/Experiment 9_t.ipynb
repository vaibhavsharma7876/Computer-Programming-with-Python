```python
# create_text_file.py
with open("sample.txt", "w") as f:
    f.write("Hello World\n")
    f.write("This\tis a\tfile\n")
    f.write("With spaces,	tabs,\n")
    f.write("and new lines.\n")

print("sample.txt created")

# count_characters.py
tabs = 0
spaces = 0
newlines = 0

with open("sample.txt", "r") as f:
    content = f.read()
    for ch in content:
        if ch == "\t":
            tabs += 1
        elif ch == " ":
            spaces += 1
        elif ch == "\n":
            newlines += 1

print("Tabs:", tabs)
print("Spaces:", spaces)
print("Newlines:", newlines)
```

```python
# create_binary_file.py
with open("source.bin", "wb") as f:
    f.write(b"This is a binary file used for testing.")

print("source.bin created")

# copy_binary.py
with open("source.bin", "rb") as src:
    data = src.read(10)

with open("destination.bin", "wb") as dest:
    dest.write(data)

print("First 10 bytes copied to destination.bin")
```
