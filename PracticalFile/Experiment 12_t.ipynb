```python
# Operator overloading using + for Student class

class Student:
    def __init__(self, name, marks):
        self.name = name
        self.marks = marks

    def __add__(self, other):
        return self.marks + other.marks


# Creating student objects
s1 = Student("Alice", 85)
s2 = Student("Bob", 90)

total_marks = s1 + s2

print("Total Marks of both students:", total_marks)
```

```python
# Voting eligibility validation

try:
    name = input("Enter your name: ")
    age = int(input("Enter your age: "))

    if age >= 18:
        print(name, "is eligible to vote.")
    else:
        print(name, "is not eligible to vote.")
except ValueError:
    print("Invalid age input. Please enter a valid integer age.")
```

```python
# Repeated integer input with exception handling

while True:
    try:
        num = int(input("Enter an integer: "))
        print("You entered:", num)
        break
    except ValueError:
        print("Invalid input, please enter an integer.")
```
