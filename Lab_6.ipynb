```python
# pretest loops
# in case of pretest loop body of loop run unconditionally for loop

i = 1
while i <= 5:
    print(i)
    i = i + 1
```

```python
# counter control. 
# no of times a loop will be executed is known in advance while in condition controlled no of times the loop is executed is not known in advance so it is a type of indefinite loop

for i in range(1, 4):
    print(i)
```

```python
week = ["Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday", "Monday"]

for i in range(31):
    print(week[i%7],  "--> ", (i+1), end='\t\t')
    if i%7 == 4:
        print(end='\n')
```

```python
for i in range(4):
    for j in range(6):
        print("*", end=" ")
    print()
```

```python
for i in range(6):
    for j in range(i):
        print("*", end=" ")
    print()
```

```python
for i in range(6):
    for j in range(6-i):
        print("*", end=" ")
    print()
```
