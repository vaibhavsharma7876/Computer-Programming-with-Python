```python
# WAP to find whether a number is positive, negative, or zero

num = float(input("Enter a number: "))

if num > 0:
    print("Positive")
elif num < 0:
    print("Negative")
else:
    print("Zero")
```

```python
# WAP to check if a character in a string is a vowel or consonant

char = input("Enter a character: ")

if ('A' <= char <= 'Z') or ('a' <= char <= 'z'):
    if char == 'A' or char == 'E' or char == 'I' or char == 'O' or char == 'U' or \
       char == 'a' or char == 'e' or char == 'i' or char == 'o' or char == 'u':
        print("Vowel")
    else:
        print("Consonant")
else:
    print("Not a valid alphabet character")
```

```python
# WAP to take input from user and check whether it is a number or character.
# If character, determine uppercase or lowercase.

ch = input("Enter a single character: ")

if '0' <= ch <= '9':
    print("It is a Number")
elif 'A' <= ch <= 'Z':
    print("It is an Uppercase Letter")
elif 'a' <= ch <= 'z':
    print("It is a Lowercase Letter")
else:
    print("It is not a number or alphabet character")
```

```python
# WAP to print largest out of three numbers

a = float(input("Enter first number: "))
b = float(input("Enter second number: "))
c = float(input("Enter third number: "))

if a >= b and a >= c:
    print("Largest is", a)
elif b >= a and b >= c:
    print("Largest is", b)
else:
    print("Largest is", c)
```

```python
# WAP to write absolute value

n = float(input("Enter Number: "))

if n < 0:
    n = -n  # multiply by -1 to make positive

print("Absolute value:", n)
```

```python
# WAP to find sum of 10 numbers

total = 0
for i in range(10):
    num = float(input(f"Enter number {i+1}: "))
    total += num

print("Sum of 10 numbers is:", total)
```

```python
# WAP to check Armstrong number (3-digit)

num = int(input("Enter a 3-digit number: "))
original = num
sum_of_cubes = 0

while num > 0:
    digit = num % 10
    sum_of_cubes += digit ** 3
    num = num // 10

if sum_of_cubes == original:
    print("Armstrong Number")
else:
    print("Not an Armstrong Number")
```

```python
# WAP to calculate factorial of a number

n = int(input("Enter number : "))
f = 1
for i in range(1, n + 1):
    f *= i

print(f)
```

```python
def fact(n):
    if n==1:
        return 1
    return n * fact(n-1)

print(fact(5))
```

```python
# WAP to print multiplication table of N

N = int(input("Enter a number: "))

for i in range(1, 11):
    print(N, "x", i, "=", N * i)
```

```python
# WAP to print leap years from 1900 to 2101

for year in range(1900, 2102):
    if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0):
        print(year)
```

```python
# WAP to sum series 1 + 1/2 + ... + 1/n

n = int(input("Enter n: "))
total = 0

for i in range(1, n + 1):
    total += 1 / i

print("Sum of series:", total)
```

```python
# WAP to sum cubes of numbers from 1 to n

n = int(input("Enter n: "))
total = 0

for i in range(1, n + 1):
    total += i ** 3

print("Sum of cubes from 1 to", n, "is:", total)
```
