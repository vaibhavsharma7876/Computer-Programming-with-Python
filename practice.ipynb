```python
def split_list(lst, n):
    result = []
    for i in range(n):
        result.append(lst[i::n])
    return result

sample = ['a','b','c','d','e','f','g','h','i','j','k','l','m','n']
print(split_list(sample, 3))
```

```python
class A:
    def show(self):
        print("A")
        
class B(A):
    def show(self):
        super().show()
        print("B")

class C(A):
    def show(self):
        super().show()
        print("C")

class D(B, C):
    def show(self):
        super().show()
        print("D")

d = D()
d.show()
```

```python
class Test:
    def __private_fun(self):
        print("This is a private function")

    def call_private(self):
        self.__private_fun()  # can call inside class
t = Test()
t.call_private()   # ✅ Works, prints: This is a private function
t.__private_fun() # ❌ Error: AttributeError
```

```python
def factorial(n):
    fact=1
    for i in range(1,n+1):
        fact = fact*i
    return fact

def sum_of_digit(n):
    n = factorial(n)

    total =0
    while n:
        total += n%10
        n = n //10

    print(total)



sum_of_digit(5)
```

```python
def is_power_of_two(n):
    return n > 0 and (n & (n - 1)) == 0

is_power_of_two(61)
```

```python
def sum_of_proper_divisors(n):
    if n == 1:
        return 0
    total = 1
    i = 2
    while i * i <= n:
        if n % i == 0:
            total += i
            if i != n // i:
                total += n // i
        i += 1
    return total


def is_abundant(n):
    return sum_of_proper_divisors(n) > n


def non_abundant_sum(limit=28123):
    # Step 1: list of abundant numbers
    abundant_numbers = [i for i in range(12, limit + 1) if is_abundant(i)]

    # Step 2: mark numbers that can be written as sum of two abundant numbers
    can_be_written = [False] * (limit + 1)

    for i in range(len(abundant_numbers)):
        for j in range(i, len(abundant_numbers)):
            s = abundant_numbers[i] + abundant_numbers[j]
            if s > limit:
                break
            can_be_written[s] = True

    # Step 3: sum numbers that cannot be written as such sums
    return sum(i for i in range(1, limit + 1) if not can_be_written[i])



result = non_abundant_sum()
print(result)
```
