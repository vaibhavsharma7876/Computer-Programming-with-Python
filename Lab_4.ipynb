```python
# WAP to increament no if positive

n = int(input("Enter Number : "))

if n >0:
    n+=1
print(n)
```

```python
# WAP to determine weather a person is eligible for vote (18+).

age = int(input("Enter age : "))

if age >= 18:
    print("Eligible")
else:
    print("Not Eligible")
```

```python
# WAP to determine character determined by user

entry = input()

if entry.isalpha():
    print("The input is alphabet")
if entry.isdigit():
    print("The input is numeric")
if entry.isspace():
    print("Empty")
```

```python
# WAP to determine weather a person is eligible to vote or not if not eligible display how many years left to be eligible.

age = int(input("Enter age : "))

if(age < 18):
    print(18 - age,"years are left to be eligible")
else :
    print("Eligible")
```

```python
# WAP to determine larger number in python

num_1 = int(input("Enter number 1 : "))
num_2 = int(input("Enter number 2 : "))

if num_1 > num_2:
    print("Larger number is:", num_1)
elif num_2 > num_1:
    print("Larger number is:", num_2)
else:
    print("Both numbers are equal")
```

```python
# WAP to enter a char and if char is in lower then convert to upper case and vice versa

user = input("Enter character : ")

if(user.lower()):
    print(user.upper())
else:
    print(user.lower())
```
