```python
a, b = 13, 6

print("Sum\t\t -->", a+b)
print("Subtraction\t -->", a-b)
print("Multiply\t -->", a*b)
print("Division\t -->", a/b)
print("Floor Division\t -->", a//b)
# a // b rounds to nearest whole number and in case of negative numbers it goes towards -ve INF.
print("Floor Division\t -->", -a//b)
```

```python
# Power / Exponent

print("b Power of a =", a**b)
print("10**-1 = ", 10** -1)
print("2** -1 = ", 2**-1)
print("7** 0 = ", 7**0)
```
