### Q1) Write a Python program to create and display all combinations of letters, selecting each letter from a different key in a dictionary.

```python
data1 = {'1': ['a', 'b'], '2': ['c', 'd']}

keys = list(data1.keys())
list1 = data1[keys[0]]
list2 = data1[keys[1]]

for x in list1:
    for y in list2:
        print(x + y)
```

### Q2) Write a Python program to get the top three items in a shop.

```python
shop = {'item1': 45.50, 'item2': 35, 'item3': 41.30, 'item4': 55, 'item5': 24}

items = list(shop.items())

for i in range(len(items)):
    max_idx = i
    for j in range(i + 1, len(items)):
        if items[j][1] > items[max_idx][1]:
            max_idx = j
    items[i], items[max_idx] = items[max_idx], items[i]

top_three = items[:3]

for item, price in top_three:
    print(item, price)
```

### Q3) Write a Python program to find the specified number of maximum values in a given dictionary.

```python
data3 = {'a': 5, 'b': 14, 'c': 32, 'd': 35, 'e': 24,
         'f': 100, 'g': 57, 'h': 8, 'i': 100}

def max_keys(data, n):
    items = list(data.items())
    
    for i in range(len(items)):
        max_idx = i
        for j in range(i + 1, len(items)):
            if items[j][1] > items[max_idx][1]:
                max_idx = j
        items[i], items[max_idx] = items[max_idx], items[i]
    
    return [key for key, value in items[:n]]

print("1 maximum:", max_keys(data3, 1))
print("2 maximum:", max_keys(data3, 2))
print("5 maximum:", max_keys(data3, 5))
```

### Q4) Write a Python program to combine two lists into a dictionary. The elements of the first one serve as keys and the elements of the second one serve as values. Each item in the first list must be unique and hashable.

```python
keys = ['a', 'b', 'c', 'd', 'e', 'f']
values = [1, 2, 3, 4, 5]

combined = {}

# Manual zip
for i in range(min(len(keys), len(values))):
    combined[keys[i]] = values[i]

print(combined)
```

### Q5) Write a Python function that evaluates mathematical expressions represented as dictionaries. The expression dictionary should support basic operations (+, -, *, /), variables, and nested expressions.

```python
def eval_expr(expr, variables={}):
    if isinstance(expr, (int, float)):
        return expr
    if isinstance(expr, str):
        return variables.get(expr, 0)

    op = expr["op"]
    left = eval_expr(expr["left"], variables)
    right = eval_expr(expr["right"], variables)

    if op == "+": return left + right
    if op == "-": return left - right
    if op == "*": return left * right
    if op == "/": return left / right

    raise ValueError("Unknown operator:", op)


# Example expression:
expression = {
    "op": "+",
    "left": {"op": "*", "left": "x", "right": 10},
    "right": 5
}

variables = {"x": 3}

print("Expression result:", eval_expr(expression, variables))
```
