```python
class student:

    name = 'StudentName'
    subject = ["GenAI", "PyhtonPrograming", "PromptEngineering"]
    marks = []

    def __init__(self, name, marks):
        self.name = name
        self.marks = marks

    def display_student(self):
        print("Name --> ", self.name)
        print("Subject\t\tMarks")
        for s, m in zip(self.subject, self.marks):
            print(f"{s}\t\t\t{m}")
            
```

```python
s = student("Divyansh", [100,100,100])
s.display_student()
```
