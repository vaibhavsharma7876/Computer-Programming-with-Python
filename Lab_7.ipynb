```python
# Break statement
for i in range(10):
    if (i==5):
        break
    print(i, end=" ")
```

```python
# Continue statement
for i in range(10):
    if i==5:
        continue
    print(i, end=" ")
```

```python
# DRY --> Dont repeat yourself.
# WET --> Write everything twice.

# WAP to define a function, using that function give diff of 2 numbers.

def diff(a, b):
    return a-b

diff(10, 11)
```

```python
# WAP to mismatch between function arguments and parameters

def addition(a, b, c):
    return a+b+c

addition(10,"hello",10)
```
